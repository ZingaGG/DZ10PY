```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
df = pd.DataFrame({'whoAmI':lst})
df.head()
```

```python
pd.get_dummies(df['whoAmI']).astype(int)
```

```python
robot = []
human = []
for elem in df['whoAmI']:
    if elem == "robot":
        robot.append("1")
        human.append("0")
    else:
        robot.append("0")
        human.append("1")    

df['robot'] = human
df['human'] = robot

print(df)
```
